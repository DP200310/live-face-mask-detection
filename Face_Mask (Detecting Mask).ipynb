```python
import cv2
from keras.models import load_model
import numpy as np
```

```python
saved_model=load_model('C:/Users/Deepanshu/Downloads/Live Face Mask Detection project/model-098.model')
```

```python
results={1:'WITHOUT MASK',0:'WITH MASK'}
bgr={1:(0,0,255),0:(0,255,0)}
rect_size = 4
video=cv2.VideoCapture(0)
font=cv2.FONT_HERSHEY_SIMPLEX
haar_data=cv2.CascadeClassifier('data.xml')
while(True):
    ret,frame=video.read()
    rerect_size = cv2.resize(frame, (frame.shape[1]//rect_size,frame.shape[0]//rect_size))
    faces = haar_data.detectMultiScale(rerect_size)
    for f in faces:
            (x,y,w,h)=[v * rect_size for v in f] 
            
            face=frame[y:y+h,x:x+w]
            predict_modified=cv2.resize(face,(150,150))
            
            nr=predict_modified/255.0
            reshaped=np.reshape(nr,(1,150,150,3))
            reshaped=np.vstack([reshaped])
            
            result=saved_model.predict(reshaped)
            label=np.argmax(result,axis=1)[0]
            
            cv2.rectangle(frame,(x,y),(x+w,y+h),bgr[label],2)
            cv2.rectangle(frame,(x,y-40),(x+w,y),bgr[label],-1)
            cv2.putText(frame,results[label], (x,y-10),font,0.8,(255,255,255),2)
    cv2.imshow("preview",frame)
    if cv2.waitKey(1) & 0XFF == ord("q"):
        break
video.release()
cv2.destroyAllWindows()
```
